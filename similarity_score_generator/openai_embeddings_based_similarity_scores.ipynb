```python
openai.api_key = ""

courses_embeddings_resp = openai.Embedding.create(
    input=courses['courses'].tolist(),
    engine="text-similarity-davinci-001")

jobs_embeddings_resp = openai.Embedding.create(
    input=courses['courses'].tolist(),
    engine="text-similarity-davinci-001")

# print(courses_embeddings_resp)
```

```python
# print(len(courses_embeddings_resp['data'][0]['embedding']))
# print(len(jobs_embeddings_resp['data'][0]['embedding']))
print(jobs_embeddings_resp['data'][0]['embedding'])
```

```python
# test single dot product of the embedding

embedding_a = courses_embeddings_resp['data'][0]['embedding']
embedding_b = jobs_embeddings_resp['data'][0]['embedding']

similarity_score = np.dot(embedding_a, embedding_b)

print(courses['courses'].tolist()[0])
print(jobs['jobs'].tolist()[0])
print(similarity_score)
```

```python
courses_str_list = courses['courses'].tolist()
jobs_str_list = jobs['jobs'].tolist()

output_df = pd.DataFrame(columns = ['course_name', 'job_name', 'similarity_score'])
# output_df.columns = ['course_name', 'job_name', 'similarity_score']

for i in range(len(courses_str_list)):
    course_em = courses_embeddings_resp['data'][i]['embedding']
    
    for j in range(len(jobs_str_list)):
        job_em = jobs_embeddings_resp['data'][j]['embedding']

        similarity_score = min(1.0, np.dot(course_em, job_em))  
#         print(courses_str_list[i])
#         print(jobs_str_list[j])
#         print(similarity_score)
        
        new_row = {'course_name': courses_str_list[i], 'job_name': jobs_str_list[j], 'similarity_score': similarity_score}
        output_df = pd.concat([output_df, pd.DataFrame([new_row])], ignore_index=True)

output_df.to_csv('sim_score_test.csv', index=False)
    
```
